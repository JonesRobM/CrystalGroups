```python
from crystallattice_toolkit.lattice import Lattice
from crystallattice_toolkit.plot.plot_atoms import plot_atoms
```

```python
lattice_vectors = [[0.5, 0.5, 0], [0.5, 0, 0.5], [0, 0.5, 0.5]]  # FCC in Å
positions = [[0, 0, 0], [0.5, 0.5, 0], [0.5, 0, 0.5], [0, 0.5, 0.5]]
species = ['Cu', 'Cu', 'Cu', 'Cu']
```

```python
lat = Lattice(lattice_vectors, species, positions)
```

```python
plot_atoms(lat, species)
```
